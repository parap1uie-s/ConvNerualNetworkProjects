```python
import csv
import pandas as pd
import numpy as np
from sklearn.utils import shuffle  
dataSrc = "data/OriginData/"

trainCSV = dataSrc + "train.csv"

destTrain = dataSrc + "../train.csv"
destTest = dataSrc + "../test.csv"
```

```python
trainData = pd.read_csv(trainCSV,header="infer",sep=',')
trainData.head()
```

```python
# 选取label=1和0的数据
print(trainData.groupby('label').count())
labelY = trainData[trainData['label'] == 1]
labelN = trainData[trainData['label'] == 0]
labelY['label'].count()
```

```python
# 抽样
labelYSample = labelY.sample(n = 2000)
labelNSample = labelN.sample(n = 4000)
labelYSample.head()
```

```python
samples = shuffle(pd.concat([labelYSample,labelNSample]))
```

```python
train = samples[["label","age","gender","education","marriageStatus","haveBaby","hometown","residence"]] .head(5000)

test = samples[["label","age","gender","education","marriageStatus","haveBaby","hometown","residence"]].tail(1000)
```

```python
train.to_csv(destTrain)
test.to_csv(destTest)
```
