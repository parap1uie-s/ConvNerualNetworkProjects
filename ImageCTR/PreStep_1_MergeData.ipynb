```python
import csv
import pandas as pd
import numpy as np

dataSrc = "data/OriginData/"
trainCSV = dataSrc + "train_ori.csv"
userCSV = dataSrc + "user.csv"

destCSV = dataSrc + "../train.csv"
```

```python
# 加载csv数据
userData = pd.read_csv(userCSV,header="infer",sep=',')
userData.head()
```

```python
trainDataOri = pd.read_csv(trainCSV,header="infer",sep=',',usecols=['label','creativeID','userID'])
trainDataOri.head()
```

```python
# Merge
trainData = pd.merge(trainDataOri, userData, on='userID')
trainData.head()
```

```python
trainData.to_csv(destCSV)
```
